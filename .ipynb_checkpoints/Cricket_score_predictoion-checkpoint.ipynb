```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')
plt.style.use("fivethirtyeight")
pd.pandas.set_option('display.max_columns',None)
```

```python
data=pd.read_csv("ipl.csv")
```

# missing values

```python
feature_with_nan=[feature for feature in data.columns if data[feature].isnull().sum()>1]
feature_with_nan
```

There is no missing values

### Categorical variable

```python
categorical_feature=[feature for feature in data.columns if data[feature].dtype=='O']
categorical_feature
```

```python
data['bat_team'].value_counts()
```

```python
len(data['bat_team'].value_counts())
```

```python
plt.figure(figsize=(30,10))
sns.countplot(data['bat_team'])
plt.show()
```

```python
plt.figure(figsize=(30,10))
sns.countplot(data['bowl_team'])
plt.show()
```

```python
for feature in categorical_feature:
    print("{} : {}".format(feature,len(data[feature].unique())))
```

### numerical variable

```python
numerical_feature=[feature for feature in data.columns if data[feature].dtype!='O']
numerical_feature
```

```python
sns.distplot(data['total'])
plt.show()
```

```python
for feature in numerical_feature:
    sns.distplot(data[feature])
    plt.show()
```
